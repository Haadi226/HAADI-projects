```python
#Q15] What is the most popular starting point for uber rides
starting_point=uber_df['START*'].dropna()
df=pd.DataFrame(starting_point.value_counts())
df.sort_values(['START*'],ascending=False)
df=df.reset_index()
df=df.rename(columns={'START*':'STARTING POINT','count':'COUNT'})
df.loc[df['COUNT']==max(df['COUNT'])]
```

```python
#Q16] What is the most popular dropping point for uber rides
dropping_point=uber_df['STOP*'].dropna()
df=pd.DataFrame(dropping_point.value_counts())
df.sort_values(['STOP*'],ascending=False)
df=df.reset_index()
df=df.rename(columns= {'STOP*':'DROPPING POINT','count':'COUNT'})
df.loc[df['COUNT']==max(df['COUNT'])]

```

```python
#17]List the most frequent route taken by drivers
df=uber_df.dropna()
df=pd.DataFrame(df.groupby(['START*','STOP*']).size())
df=df.rename(columns={0:'Count'})
df=df.sort_values(['Count'],ascending=False)
df.loc[df['Count']==max(df['Count'])]
```

```python
#18]Print all the purpose of the trips
print(np.array(uber_df['PURPOSE*'].dropna().unique()))
uber_df['MILES*'].groupby(uber_df['PURPOSE*']).sum()
```

```python
#Plot a bar graph of purpose vs Distance
df=pd.DataFrame(uber_df['MILES*'].groupby(uber_df['PURPOSE*']).sum())
df.plot(kind="bar")
plt.bar(df.index,df['MILES*'],color='green',alpha=0.7)
plt.xlabel("Purpose")
plt.ylabel("Distance")
plt.title("Purpose vs Distance")
plt.legend()
plt.show()
```

```python
#pie chart based on category

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

uber_df=pd.read_csv('/content/uberdrive.csv')

# ... (previous code)

#Plot number of trips vs category of trips
category_counts = uber_df['CATEGORY*'].value_counts()

# Create the pie chart
plt.figure(figsize=(5,5))  # Adjust figure size as needed
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Uber Trip Categories')
plt.legend()
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
